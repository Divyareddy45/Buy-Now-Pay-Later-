```python
data['transaction_date'] = pd.to_datetime(data['transaction_date'])
data['account_created'] = pd.to_datetime(data['account_created'])
```

```python
data['total_financed_amount'] = data['purchase_amount'] - data['down_payment']
```

```python
data['total_repayment'] = data['down_payment'] + (data['installment_amount'] * data['installments'])
```

```python
data['repayment_ratio'] = data['total_repayment'] / data['purchase_amount']
```

```python
data['high_risk'] = ((data['repayment_ratio'] < 0.8) | (data['late_payments'] >= 2) | (data['credit_score'] < 550)).astype(int)
print(data['high_risk'])
```
